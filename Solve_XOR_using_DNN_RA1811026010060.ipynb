# Import Libraries

These are all the libraries we need to run our code. 
* One of the most popular libraries is **numpy**. It provides strong support for python to work with arrays. We import numpy and alias it as np which is pretty common thing to do.
* **Keras** is an open-source software library that provides a Python interface for **artificial neural networks**. Keras acts as an interface for the TensorFlow library.
* Keras offers two different APIs to construct a model: **a functional and a sequential** one. We’re using the sequential API hence the second import of Sequential from keras.models.

```
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

# Defining Training and test data

* We initialize **training_data** as a two-dimensional array (an array of arrays) where each of the inner arrays has exactly two items.
* We setup **target_data** as another two-dimensional array. All the inner arrays in target_data contain just a single item though.
* Each inner array of training_data relates to its counterpart in target_data.

Truth-table-of-the-logical-operator-XOR.png

* From the above table we infer that **x1, x2 comprise the training data** and **x1 XOR x2 comprises the target_data.**

```
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]], "float32")
```

# Define the Model

* The first line sets up an empty model using the **Sequential API**.
* We’re adding a **Dense layer** to our model.
* We set **input_dim=2** because each of our input samples is an array of length 2 ([0, 1], [1, 0] etc.).
* The **16** stands for the **dimension of the output** for this layer.
* In terms of neurons it means that we have two input neurons (input_dim=2) spreading into 16 neurons in a so called **hidden layer**.
* We also added another layer with an output dimension of 1 and without an explicit input dimension.
* We set the activation function for **input layer as ReLu** and **hidden layer as Sigmoid.**

xor_model.png

```
model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

* We use the **mean_squared_error** as the loss for model.
* The job of the optimizer is it to find the right adjustments for the weights. We use **Adam** as the optimizer for the model.
* We set **binary_accuracy** as the metric for our model. The metric gives us access to a number that tells us exactly how accurate our predictions are.

```
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

* The first two params are **training and target data.**
* The third one is the number of **epochs (learning iterations)**.
* The last one tells keras how much info to print out during the training.

```
model.fit(training_data, target_data, epochs=500, verbose=2)
```

* Once the training phase finished we can start making predictions.
* In real world scenarios, the data used for training the model is never used for testing it. 
* In other words, we check if the model works well with the new data or  does it just memorize all the data and expected results it had seen in the training phase.

**Note**
> * We are rounding the output to get clear binary answers. 
*  Neural networks calculate probabilities. If we wouldn’t round we would see something like 0.9993... and 0.00034... instead of 1 and 0.

```
print(model.predict(training_data).round())
```

# Make Changes to the model

```
model_1 = Sequential()
model_1.add(Dense(32, input_dim=2, activation='relu'))
model_1.add(Dense(1, activation='sigmoid'))
```

```
model_1.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

**Note**
> We see that from changing the dimensions of hidden layer from 16 to 32 we attain 100% accuracy in just **38 epochs**.

```
model_1.fit(training_data, target_data, epochs=500, verbose=2)
```

```
model_2 = Sequential()
model_2.add(Dense(32, input_dim=2, activation='relu'))
model_2.add(Dense(32, activation='relu'))
model_2.add(Dense(1, activation='sigmoid'))
```

```
model_2.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

**Note**
> We note that by stacking an **additional hidden layer** with 32 units we attain 100% accuracy in just **17 epochs.**

```
model_2.fit(training_data, target_data, epochs=100, verbose=2)
```
